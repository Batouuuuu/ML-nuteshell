Préparer les données 

```python
new_df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
    #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #'Track layout', 'Fastrack available', 'Soft opening date.1',
    # 'Closing date', 
    # 'Opened', 
    #'Replaced by', 'Website',
    # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    # 'Single rider line available', 'Restraint Style',
    # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
    'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
    'speed_mph', 
    # 'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()  ## copy permet d'indiquer à python qu'on a nouveau dataframe et qu'on ne se refère plus à l'ancien
```

```python
new_df.shape
```

```python
# exemple pour drop une seule colonne
# df.drop(['Opening date'], axis=1)
```

```python
# changer le type d'une colonne on passe d'un object (donc du texte) à une date
new_df['opening_date_clean'] = pd.to_datetime(new_df['opening_date_clean'])
# il y'a aussi pd.to_numeric
```

```python
#renomer une colonne 
new_df = new_df.rename(columns={"coaster_name":"Coaster_Name",
                       "year_introduced":"Year_Introduced",
                        "opening_date_clean":"Opening_Date",
                        "speed_mph":"Speed_mph",
                        "height_ft": "Height_ft",
                        "Inversions_clean": "Inversions",
                        "Gforce_clean": "Gforce" })
```

```python
## vérifier un exemple duppliqué 
new_df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
#new_df.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"]).sum()
new_df = new_df[~new_df.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"])] \
    .reset_index(drop=True).copy()
```

Feature understanding (univariate analysis)

```python
ax = new_df["Year_Introduced"].value_counts().head(10) \
.plot(kind="bar", title="Top 10 years Coasters Introduced")

ax.set_xlabel("Year Introduced")
ax.set_ylabel("Count")
```

```python
new_df["Speed_mph"].plot(kind="hist", 
                         bins=20, 
                         title= "Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
```

```python
new_df["Speed_mph"].plot(kind="kde", 
                         title= "Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
```

RELATIONS ENTRE LES FEATURES

```python
new_df.plot(kind="scatter", x="Speed_mph", y="Height_ft", title="Coaster speed vs height")
plt.show() ## façon pour ne pas afficher l'object
```

```python
sns.scatterplot(x="Speed_mph", y="Height_ft",
                hue="Year_Introduced", data=new_df)
```

```python
## si on veut comparer plus de deux features
sns.pairplot(new_df, vars=["Year_Introduced", "Speed_mph", "Height_ft","Inversions","Gforce"], hue="Type_Main")
plt.show()
```

```python
## correlation 
df_cor = new_df[["Year_Introduced", "Speed_mph", "Height_ft","Inversions","Gforce"]].dropna().corr()
df_cor
```

```python
sns.heatmap(df_cor, annot=True)
```

Questions sur les datas

```python
## Quelles sont les locations avec les rollers coaster les plus rapide (minimum 10 )
new_df["Location"].value_counts()
```

```python
## on se rend compte que Other ne compte pas comme un pays donc on va l'exclure avec une query
ax = new_df.query('Location != "Other"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean', 'count']) \
.query('count >= 10') \
.sort_values('mean')['mean'] \
.plot(kind="barh", figsize=(12, 5), title="Average Coast Speed by Location")

ax.set_xlabel("Average Coaster Speed")
plt.show()
```
